```python
import numpy as np
import pandas as pd
from numpy import random
import os
os.getcwd()
os.chdir("/content/drive/MyDrive")
os.getcwd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

```

```python
df=pd.read_excel("loan_logistic_regression_dataset.xlsx")
df.head()
```

```python
df = pd.get_dummies(df,columns=["Gender","Education_Level"],drop_first=True).astype(int)
```

```python
df.head()
```

```python
x = df.drop("Loan_Approved", axis=1)

y = df["Loan_Approved"]
```

```python

x_train, x_test, y_train, y_test = train_test_split( x, y,test_size=0.20,random_state=20)
```

```python
x_train=sm.add_constant(x_train,prepend=False)
x_test=sm.add_constant(x_test,prepend=False)
```

```python

x_train.head()
```

```python
model = sm.Logit(y_train, x_train)
# result = model.fit(method="bfgs", maxiter=500)
result = model.fit_regularized(method='l1', alpha = 0.1)

print(result.summary())
```
